### <font, color="Blue"/> Replacing spaces in category names with hyphens and converting them to lower case

```python
def insertHyphen(catName):
    if len(catName.split()) > 1:
        return ('_').join(catName.split()).lower()
    else:
        return catName.lower()
```

```python
prodLookup['category'] = prodLookup['category'].apply(insertHyphen)
```

# Download images to disk

```python
pidList = prodLookup['_unit_id'].tolist()
```

```python
def getImages(df,urlCol,rowIndex,imageList,fpath):
    
    for each_pid in imageList:
        #get the image url
        uStr = df[urlCol][df[rowIndex]==each_pid]
        
        #convert Series to list and retrieve the first element (URL string)
        uStr = uStr.tolist()[0]
        
        #construct filename <category>-<_unit_id>
        imgName = df[urlCol][df[rowIndex]==each_pid]+'-'+str(each_pid)+'.jpg'
        
        #save image to disk
        urllib.request.urlretrieve(url=uStr, filename=fpath+imgName)
```

```python
def getImages(df,urlCol,prodCol,labelCol,fpath,start=0,stop=1200):
    
    for each in df.index[start:stop]:
        #get the image url
        uStr = str(df[urlCol][each])
        
        #construct filename <category>-<_unit_id>
        imgName = df[labelCol][each]+'-'+str(df[prodCol][each])+'.jpg'
        
        #save image to disk
        urllib.request.urlretrieve(url=uStr, filename=fpath+imgName)
    
    print('Download complete')
    print('Last row index = {%d}'.format({each}))
```

```python
getImages(df=prodLookup,urlCol='image_url',prodCol='_unit_id',labelCol='category',fpath=base+img,start=1129,stop=None)
```
